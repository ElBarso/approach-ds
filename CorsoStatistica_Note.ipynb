```python
H0 = 6
#data = np.array([-599,-498,3,4,505,606])
data2 = np.array([3.5,3.5,3,3.5,4,3.5])
d = 0
data2[0] = data2[0] - d
data2[5] = data2[5] + d
n = len(data2) - 1 

t_mio = (np.mean(data2) - H0) / (np.std(data2) / np.sqrt(n) )
print(t_mio)

t, p = stats.ttest_1samp(data2, H0) 
print(t, p)

x = np.linspace(-0, 9, 101)
m = np.mean(data2)
s = np.std(data2)
plt.plot(x, stats.norm.pdf(x,6), x, stats.norm.pdf(x, m, s))
c = m
plt.plot([c,c],[0,0.4],'r--')
plt.legend(['H_0', 'H_a'])
```
