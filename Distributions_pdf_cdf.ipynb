```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
from scipy.stats import norm
```

```python
## Gaussian

# number of point
N = 1001

x = np.linspace(-4,4,N)
gauss = stats.norm.pdf(x)

## Sample generation from gaussian distribution

# distribution parameters
std     = 1    # stretch
mean    = 5    # shift distribution orizontally
n       = 1000 # number of points

# data generation
sample = std*np.random.randn(n) + mean

# plot
fig, (ax1, ax2, ax3) = plt.subplots(1,3,figsize=(13.5,3.5))
ax1.plot(x,gauss)
ax1.set_title('Analytic Gaussian (normal) distribution')
ax2.hist(sample, bins=30)
ax2.set_title('Sampled data from Gaussian distribution')
ax3.hist(norm.pdf(np.linspace(-4, 4, n)), bins=30)
ax3.set_title('Data histogram from analytic Gaussian distribution')
plt.tight_layout()
plt.show()
```

```python
## t-student

# number of point
N   = 1001
dof = 2

x = np.linspace(-4,4,N)
t = stats.t.pdf(x, dof)

## Sample generation from t distribution

# distribution parameters
width     = 1    # stretch
center    = 50    # shift distribution orizontally
n         = 1000 # number of points

# data generation
sample = width*np.random.standard_t(dof, n) + center

# plot
fig, (ax1, ax2, ax3) = plt.subplots(1,3,figsize=(13.5,3.5))
ax1.plot(x,t)
ax1.set_title('Analytic t-student distribution')
ax2.hist(sample, bins=30)
ax2.set_title('Sampled data from t-student distribution')
ax3.hist(stats.t.pdf(np.linspace(-4, 4, n), dof), bins=30)
ax3.set_title('Data histogram from analytic t-student distribution')
plt.tight_layout()
plt.show()
```

```python
## log-norm

# number of point
N   = 1001

x = np.linspace(-4,4,N)
lognorm = stats.lognorm.pdf(x, 2)

## Sample generation from log-normal distribution

# distribution parameters
sigma = 1    # stretch
mean  = 0    # shift distribution orizontally
n     = 1000 # number of points

# data generation
sample = np.random.lognormal(mean,sigma,n)

# plot
fig, (ax1, ax2, ax3) = plt.subplots(1,3,figsize=(13.5,3.5))
ax1.plot(x,lognorm)
ax1.set_title('Analytic Log-normal distribution')
ax2.hist(sample, bins=30)
ax2.set_title('Sampled data from Log-normal distribution')
ax3.hist(stats.lognorm.pdf(np.linspace(-4, 4, n), 2), bins=30)
ax3.set_title('Data histogram from analytic Log-normal distribution')
plt.tight_layout()
plt.show()
```

```python
from scipy import stats as s

fig, ax = plt.subplots(1,2,figsize=(16,6))

x1 = np.linspace(-3, 3, 101)
x2 = np.linspace(-10, 10, 101)

p1 = s.logistic.pdf(x1)
p2 = s.norm.pdf(x1)
p3 = s.lognorm.pdf(x1,1)
p4 = s.chi2.pdf(x1,1)
p5 = s.exponnorm.pdf(x1,1)

c1 = s.logistic.cdf(x2)
c2 = s.norm.cdf(x2)
c3 = s.lognorm.cdf(x2,1)
c4 = s.chi2.cdf(x2,1)
c5 = s.exponnorm.cdf(x2,1)

ax[0].plot(x1, p1, x1, p2, x1, p3, x1, p4, x1, p5, lw=3)
ax[0].set_title("pdf of some distributions")
ax[0].legend(['logistic', 'gaussian', 'lognorm','chi$^2$','exponnorm'])
ax[1].plot(x2, c1, x2, c2, x2, c3, x2, c4, x2, c5, lw=3)
ax[1].set_title("cdf of some distributions")
ax[1].legend(['logistic', 'gaussian', 'lognorm','chi$^2$','exponnorm'])
```

```python
## binomial
n = 10 # number die tosses
p = .5 # probability of even number
xb = range(n+1)
bindist = stats.binom.pmf(xb,n,p)

## Gaussian
N = 1001  # number of point
xg = np.linspace(-4,4,N)
gauss = stats.norm.pdf(xg)

# plot
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,4))
ax1.bar(xb,bindist)
ax1.set_title('Binomial distribution (n=%s, p=%g)'%(n,p))
shift = 176
ax2.plot(xg + shift,gauss)
x_area = np.linspace(-1,-0.5,1001)
y_area = stats.norm.pdf(x_area)
ax2.fill_between(x_area + shift, y_area, step="pre", alpha=0.4)
ax2.set_title('Gaussian (normal) distribution')
plt.tight_layout()
plt.show()
```

```python

## Uniformly-distributed numbers

# parameters
stretch = 2 # not the variance
shift   = .5
n       = 10000

# create data
data = stretch*np.random.rand(n) + shift-stretch/2

# plot data
fig,ax = plt.subplots(2,1,figsize=(5,6))

ax[0].plot(data,'.',markersize=1)
ax[0].set_title('Uniform data values')

ax[1].hist(data,25)
ax[1].set_title('Uniform data histogram')

plt.show()
```

```python
## log-normal distribution

N = 1001
x = np.linspace(0,10,N)
lognormdist = stats.lognorm.pdf(x,1)

plt.plot(x,lognormdist)
plt.title('Analytic log-normal distribution')
plt.show()
```

```python
## empirical log-normal distribution

shift   = 5  # equal to the mean?
stretch = .5 # equal to standard deviation?
n = 2000     # number of data points

# generate data
data = stretch*np.random.randn(n) + shift
data = np.exp( data )

# plot data
fig,ax = plt.subplots(2,1,figsize=(4,6))
ax[0].plot(data,'.')
ax[0].set_title('Log-normal data values')

ax[1].hist(data,25)
ax[1].set_title('Log-normal data histogram')
plt.show()
```

```python
## binomial

# a binomial distribution is the probability of K heads in N coin tosses,
# given a probability of p heads (e.g., .5 is a fair coin).

n = 10 # number on coin tosses
p = .5 # probability of heads

x = range(n+2)
bindist = stats.binom.pmf(x,n,p)

plt.bar(x,bindist)
plt.title('Binomial distribution (n=%s, p=%g)'%(n,p))
plt.show()
```

```python
## t

x  = np.linspace(-4,4,1001)
df = 200
t  = stats.t.pdf(x,df)

plt.plot(x,t)
plt.xlabel('t-value')
plt.ylabel('P(t | H$_0$)')
plt.title('t(%g) distribution'%df)
plt.show()
```

```python
## F

# parameters
num_df = 5   # numerator degrees of freedom
den_df = 100 # denominator df

# values to evaluate 
x = np.linspace(0,10,10001)

# the distribution
fdist = stats.f.pdf(x,num_df,den_df)

plt.plot(x,fdist)
plt.title(f'F({num_df},{den_df}) distribution')
plt.xlabel('F value')
plt.show()
```
